```python
# all_data['Last Name'] = all_data['Name'].str[-1:]
def get_city(x):
    l=len(x.split(' '))
    if l <= 2:
        return x.split(' ')[1]
    return x.split(' ')[2]

def get_state(y):
    return y.split(' ')[0]

#all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ '(' + get_state(x) + ')')
#doing same thing as above i have did with f string function
all_data['Last Name'] = all_data['Name'].apply(lambda x: f"{get_city(x)}")
all_data['First Name'] = all_data['Name'].apply(lambda y: f"{get_state(y)}")
#all_data.drop(columns = 'State', inplace=True)

# all_data
```
