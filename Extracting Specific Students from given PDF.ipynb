```python
m_name = pd.read_excel("D:\Coding Boot Camp\muslim-names11.xlsx", engine='openpyxl')
m_name = m_name.dropna(how='all')
# m_name.head()

name_list = m_name["Name"].tolist()
muslim_name = []
raw_name=["Aman","Amber","Sham","Diya","Rani","Roshan","Gulab","Naman","Sahil","Seema","Sameer","Uday","Sikandar","Samir"]
commom_name = []
for j in raw_name:
    b = j.upper()
    commom_name.append(b)
print(commom_name)
for i in name_list:
    if len(i) > 2:
        a = i.upper()
        if a not in commom_name:
            muslim_name.append(a)
# print(muslim_name)
```

```python
#Surname
m_surname = pd.read_excel("D:\Coding Boot Camp\surname.xlsx", engine='openpyxl')
m_surname.head()
surname_list = m_surname["Surname"].tolist()
muslim_surname = []
for i in surname_list:
    a = i.upper()
    if a not in commom_name:
        muslim_surname.append(a)

all_name=muslim_surname+muslim_name
"SHAM" in muslim_surname
```

```python
def get_city(x):
    l=len(x.split(' '))
    if l > 4:
        return x.split(' ')[3]
    elif l >= 3 and l<4:
        return x.split(' ')[2]
    elif l >= 2 and l<3:
        return x.split(' ')[1]
    return x.split(' ')[0]

def get_state(y):
    return y.split(' ')[0]


all_df_data['Last Name'] = all_df_data['Name'].apply(lambda x: f"{get_city(x)}")
all_df_data['First Name'] = all_df_data['Name'].apply(lambda y: f"{get_state(y)}")
all_df_data
```

```python
bucket=[]

for k in all_name:
    #ln
    ln_df = all_df_data[all_df_data['Last Name'] == k]
    if not ln_df.empty:
        bucket.append(ln_df)
    #fn
    fn_df = all_df_data[all_df_data['First Name'] == k]
    if not fn_df.empty:
        bucket.append(fn_df)
```

```python
len(bucket)
bucket[0]
```

```python
df_results = pd.concat(bucket)
results = df_results.drop_duplicates()
```
