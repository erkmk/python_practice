```python
bucket_attr = [
    "enabled",
    "Group Key",
    "Unit Threshold",
]

formula_attr_1 = [
    "Reportable Name",
    "Use CPPD",
    "Allow Overlap",
]
```

```python
import openpyxl
wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = "PriceType Version Comparision"

price_type = 'amp'
v1='2'
v2='6'
header = {
    price_type +'-v'+v1:"", #4
    price_type +'-v'+v2:"", #5 
    "match": "" #6
}

sheet.cell(row=2, column=1).value = 'bucket'

for y, col in enumerate(header, start=4):
    sheet.cell(row=1, column=y).value = col

buckets={
        "rebates": {
            "enabled": {
                "v1": True,
                "v2": True,
                "match": True
            },
            "Group Key": {
                "v1": None,
                "v2": None,
                "match": True
            },
            "Unit Threshold": {
                "v1": "0.0000",
                "v2": "0.0000",
                "match": True
            },
        }
}

formulas ={"rcp_sales":{
                "Reportable Name": {
                "v1": "Ineligible Rebates - State",
                "v2": "Ineligible Rebates - State",
                "match": True
                },
                "Use CPPD": {
                    "v1": False,
                    "v2": False,
                    "match": True
                },
                "Allow Overlap": {
                    "v1": False,
                    "v2": False,
                    "match": True
                },
}
}
row_position=3
for bucket_name in buckets:
    sheet.cell(row=row_position, column=2).value = bucket_name
    for k, attr in enumerate(bucket_attr, start=row_position+1):
        print(attr)
        data=buckets[bucket_name]
        sheet.cell(row=k, column=3).value = attr
        sheet.cell(row=k, column=4).value = "" if data.get(attr)['v1']==None else data.get(attr)['v1']
        sheet.cell(row=k, column=5).value = "" if data.get(attr)['v2']==None else data.get(attr)['v2']
        sheet.cell(row=k, column=6).value = 'Y' if data.get(attr)['match'] else 'N'
        row_position=k
    row_position +=1
    
print("row_position-bucket", row_position)

row_position += 1
sheet.cell(row=row_position, column=1).value = 'formula'
row_position += 1
for formula_name in formulas:
    print("formula_name", formula_name)
    sheet.cell(row=row_position, column=2).value = formula_name
    for k, attr in enumerate(formula_attr_1, start=row_position+1):
        print(attr)
        data=formulas[formula_name]
        sheet.cell(row=k, column=3).value = attr
        sheet.cell(row=k, column=4).value = "" if data.get(attr)['v1']==None else data.get(attr)['v1']
        sheet.cell(row=k, column=5).value = "" if data.get(attr)['v2']==None else data.get(attr)['v2']
        sheet.cell(row=k, column=6).value = 'Y' if data.get(attr)['match'] else 'N'
        row_position=k
    row_position +=1

wb.save('test1.xlsx')
print(row_position)
```
